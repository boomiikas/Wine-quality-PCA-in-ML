```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
plt.figure(figsize=(15,8))
for i,cols in enumerate(df.columns,1):
    plt.subplot(3, 4, i)
    sns.histplot(df[cols],kde=True,fill=True)
    plt.title(cols)
plt.tight_layout()
plt.show()
```

```python
df.skew()
```

```python
df['residual sugar']=np.log1p(df['residual sugar'])
df['chlorides']=np.log1p(df['chlorides'])
df['total sulfur dioxide']=np.log1p(df['total sulfur dioxide'])
df['free sulfur dioxide']=np.log1p(df['free sulfur dioxide'])
df['sulphates']=np.log1p(df['sulphates'])
df['alcohol']=np.log1p(df['alcohol'])
df.skew()
```

```python
plt.figure(figsize=(15,8))
for i,cols in enumerate(df.columns,1):
    plt.subplot(3, 4, i)
    sns.histplot(df[cols],kde=True,fill=True)
    plt.title(cols)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,8))
for i,cols in enumerate(df.columns,1):
    plt.subplot(3,4,i)
    plt.hist(df[cols],color='lightgreen',bins=10,edgecolor='black')
    plt.title(cols)
plt.tight_layout()
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="Blues")
plt.show()
```

```python
plt.figure(figsize=(15,8))
for i,cols in enumerate(df.columns,1):
    plt.subplot(3,4,i)
    plt.boxplot(df[cols])
    plt.title(cols)
plt.tight_layout()
plt.show()
```
